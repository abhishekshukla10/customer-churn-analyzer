Find Best Threshold Systematically

Instead of manually trying 0.4, 0.3, etc., well:

Evaluate multiple thresholds

Track recall, precision, F1

Choose threshold based on business objective

```python
#Step 1  Generate Threshold Candidates

import numpy as np
from sklearn.metrics import precision_score, recall_score, f1_score

thresholds = np.arange(0.1, 0.9, 0.05)

results = []

for t in thresholds:
    y_pred_t = (y_prob >= t).astype(int)
    
    precision = precision_score(y_test, y_pred_t)
    recall = recall_score(y_test, y_pred_t)
    f1 = f1_score(y_test, y_pred_t)
    
    results.append((t, precision, recall, f1))

results_df = pd.DataFrame(results, columns=["Threshold", "Precision", "Recall", "F1"])
results_df
```

```python
import shap
import matplotlib.pyplot as plt
import numpy as np

# ============================================
# WORKAROUND: Manual Summary Plot
# ============================================

def plot_shap_summary_manual(shap_values, X_test):
    """
    Manual SHAP summary plot to avoid NumPy compatibility issues
    """
    # Calculate mean absolute SHAP values
    mean_abs_shap = np.abs(shap_values.values).mean(axis=0)
    
    # Create dataframe
    feature_importance = pd.DataFrame({
        'feature': X_test.columns,
        'importance': mean_abs_shap
    }).sort_values('importance', ascending=True)
    
    # Plot
    plt.figure(figsize=(10, 8))
    plt.barh(feature_importance['feature'], feature_importance['importance'], 
             color='steelblue', alpha=0.7)
    plt.xlabel('Mean |SHAP Value| (Average Impact on Model Output)', fontsize=12)
    plt.title('SHAP Feature Importance', fontsize=14, weight='bold')
    plt.tight_layout()
    plt.show()

# Use the workaround
plot_shap_summary_manual(shap_values, X_test)
```

```python
# MOdel Summary Dashboard

import matplotlib.pyplot as plt
from sklearn.metrics import roc_auc_score, confusion_matrix
import seaborn as sns

def create_interpretability_dashboard(model, X_test, y_test, X_train):
    """
    Create comprehensive model interpretability dashboard
    """
    # Get predictions
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1]
    
    # Create figure
    fig = plt.figure(figsize=(20, 12))
    
    # 1. Feature Importance
    ax1 = plt.subplot(3, 3, 1)
    coefficients = pd.Series(model.coef_[0], index=X_train.columns).sort_values(ascending=False)
    top_10 = coefficients.head(10)
    ax1.barh(range(len(top_10)), top_10.values, color='steelblue')
    ax1.set_yticks(range(len(top_10)))
    ax1.set_yticklabels(top_10.index)
    ax1.set_xlabel('Coefficient')
    ax1.set_title('Top 10 Feature Importance', weight='bold')
    ax1.invert_yaxis()
    
    # 2. Confusion Matrix
    ax2 = plt.subplot(3, 3, 2)
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax2)
    ax2.set_xlabel('Predicted')
    ax2.set_ylabel('Actual')
    ax2.set_title('Confusion Matrix', weight='bold')
    
    # 3. ROC Curve
    from sklearn.metrics import roc_curve
    ax3 = plt.subplot(3, 3, 3)
    fpr, tpr, _ = roc_curve(y_test, y_prob)
    auc = roc_auc_score(y_test, y_prob)
    ax3.plot(fpr, tpr, label=f'AUC = {auc:.3f}')
    ax3.plot([0, 1], [0, 1], 'k--')
    ax3.set_xlabel('False Positive Rate')
    ax3.set_ylabel('True Positive Rate')
    ax3.set_title('ROC Curve', weight='bold')
    ax3.legend()
    
    # 4. Probability Distribution
    ax4 = plt.subplot(3, 3, 4)
    ax4.hist(y_prob[y_test==0], bins=30, alpha=0.5, label='No Churn', color='green')
    ax4.hist(y_prob[y_test==1], bins=30, alpha=0.5, label='Churn', color='red')
    ax4.axvline(0.5, color='black', linestyle='--')
    ax4.set_xlabel('Predicted Probability')
    ax4.set_ylabel('Count')
    ax4.set_title('Prediction Distribution', weight='bold')
    ax4.legend()
    
    # 5. Calibration Curve
    from sklearn.calibration import calibration_curve
    ax5 = plt.subplot(3, 3, 5)
    prob_true, prob_pred = calibration_curve(y_test, y_prob, n_bins=10)
    ax5.plot(prob_pred, prob_true, marker='o')
    ax5.plot([0, 1], [0, 1], 'k--')
    ax5.set_xlabel('Predicted Probability')
    ax5.set_ylabel('True Probability')
    ax5.set_title('Calibration Curve', weight='bold')
    
    # 6. Feature Correlation
    ax6 = plt.subplot(3, 3, 6)
    feature_corr = pd.DataFrame({
        'feature': X_train.columns,
        'correlation': [X_train[col].corr(pd.Series(y_train)) for col in X_train.columns]
    }).sort_values('correlation', key=abs, ascending=False).head(10)
    ax6.barh(range(len(feature_corr)), feature_corr['correlation'].values)
    ax6.set_yticks(range(len(feature_corr)))
    ax6.set_yticklabels(feature_corr['feature'])
    ax6.set_xlabel('Correlation')
    ax6.set_title('Feature-Target Correlation', weight='bold')
    ax6.invert_yaxis()
    
    # 7. Metrics Summary
    from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
    ax7 = plt.subplot(3, 3, 7)
    metrics = {
        'Accuracy': accuracy_score(y_test, y_pred),
        'Precision': precision_score(y_test, y_pred),
        'Recall': recall_score(y_test, y_pred),
        'F1-Score': f1_score(y_test, y_pred),
        'AUC': auc
    }
    ax7.barh(list(metrics.keys()), list(metrics.values()), color='coral')
    ax7.set_xlim([0, 1])
    ax7.set_xlabel('Score')
    ax7.set_title('Performance Metrics', weight='bold')
    for i, (k, v) in enumerate(metrics.items()):
        ax7.text(v + 0.02, i, f'{v:.3f}', va='center')
    
    # 8. Top Positive Features
    ax8 = plt.subplot(3, 3, 8)
    top_pos = coefficients[coefficients > 0].head(5)
    ax8.barh(range(len(top_pos)), top_pos.values, color='red', alpha=0.7)
    ax8.set_yticks(range(len(top_pos)))
    ax8.set_yticklabels(top_pos.index)
    ax8.set_xlabel('Coefficient')
    ax8.set_title('Top Churn Drivers', weight='bold')
    ax8.invert_yaxis()
    
    # 9. Top Negative Features
    ax9 = plt.subplot(3, 3, 9)
    top_neg = coefficients[coefficients < 0].tail(5).sort_values()
    ax9.barh(range(len(top_neg)), top_neg.values, color='green', alpha=0.7)
    ax9.set_yticks(range(len(top_neg)))
    ax9.set_yticklabels(top_neg.index)
    ax9.set_xlabel('Coefficient')
    ax9.set_title('Top Retention Factors', weight='bold')
    ax9.invert_yaxis()
    
    plt.suptitle('Model Interpretability Dashboard', fontsize=20, weight='bold', y=0.995)
    plt.tight_layout()
    plt.show()

# Create dashboard
create_interpretability_dashboard(model_2, X_test, y_test, X_train)
```

##Proper Logistic Optimization

Why Regularization Matters

Logistic regression uses regularization by default:

Default = L2 regularization

Strength controlled by parameter C

Important:

Smaller C  stronger regularization

Larger C  weaker regularization

Regularization prevents:

Overfitting

Extreme coefficients

Instability

```python
#Proper Logistic Optimization

#Standardize Features (Very Important) Add scaling

from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
```

```python
#Build Pipeline
pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("logreg", LogisticRegression(
        max_iter=2000,
        class_weight='balanced'
    ))
])
```

```python
#Hyperparameter Tuning (C Parameter)

from sklearn.model_selection import GridSearchCV

param_grid = {
    "logreg__C": [0.01, 0.1, 1, 5, 10]
}

grid = GridSearchCV(
    pipeline,
    param_grid,
    cv=5,
    scoring="roc_auc",
    n_jobs=-1
)

grid.fit(X_train, y_train)
```

```python
# Apply business filtering

# Get contract type from one-hot encoded columns
def get_contract_type(row):
    if 'Contract_Two year' in X_test.columns and row['Contract_Two year'] == 1:
        return "Two year"
    elif 'Contract_One year' in X_test.columns and row['Contract_One year'] == 1:
        return "One year"
    else:
        return "Month-to-month"
# Create complete analysis for all customers
results = []

for i in range(len(y_prob_best)):
    prob = y_prob_best[i]
    contract = get_contract_type(X_test.iloc[i])
    risk = classify_risk(prob)
    
    if contract in ["One year", "Two year"]:
        action = "No immediate action - contract protected"
    elif risk == "High":
        action = "Immediate intervention required"
    elif risk == "Medium":
        action = "Proactive engagement"
    elif risk == "Low":
        action = "Standard monitoring"
    else:
        action = "No action needed"
    
    results.append({
        'Customer': i,
        'Probability': prob,
        'Risk': risk,
        'Contract': contract,
        'Action': action
    })

results_df = pd.DataFrame(results)
print(results_df.head())
```
